```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV (reemplaza 'ruta_del_archivo.csv' por la ruta de tu archivo)
df = pd.read_csv('archivo_limpio.csv')

# Convertir la columna 'fecha' en tipo datetime para mejor manejo en la gráfica
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d_%H-%M-%S')

# Crear la gráfica
plt.figure(figsize=(10,6))
plt.plot(df['fecha'], df['valor_corriente'], marker='o')

# Añadir etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Valor Corriente (A)')
plt.title('Gráfico de Corriente en función del Tiempo')

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.tight_layout()  # Para ajustar los márgenes
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV (reemplaza 'ruta_del_archivo.csv' por la ruta de tu archivo)
df = pd.read_csv('archivo_limpio.csv')

# Convertir la columna 'fecha' a formato datetime para mejor manejo
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d_%H-%M-%S')

# Definir el rango de fechas para filtrar (ajusta las fechas según necesites)
fecha_inicio = '2024-09-11 10:54:00'
fecha_fin = '2024-09-11 11:08:00'

# Filtrar el DataFrame para que contenga solo los datos dentro del rango de fechas
df_filtrado = df[(df['fecha'] >= fecha_inicio) & (df['fecha'] <= fecha_fin)]

# Crear la gráfica con el rango de fechas seleccionado
plt.figure(figsize=(10,6))
plt.plot(df_filtrado['fecha'], df_filtrado['valor_corriente'], marker='o')

# Añadir etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Valor Corriente (A)')
plt.title(f'Gráfico de Corriente desde {fecha_inicio} hasta {fecha_fin}')

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.tight_layout()  # Para ajustar los márgenes
plt.show()
```
