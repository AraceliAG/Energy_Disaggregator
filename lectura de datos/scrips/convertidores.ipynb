AJUSTAR FORMATO DE FECHAS 

```python
#AJUSTA EL FORMATO DE FECHA DE UN ARCHIVO CSV


import pandas as pd

# Ruta del archivo CSV que deseas modificar
ruta_archivo = 'C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04/Corriente_Avg/archivo_concatenado.csv'

# Especificar el formato original de las fechas y el formato deseado
formato_original = '%Y/%m/%d %H:%M:%S'  # Cambia esto por el formato real
formato_deseado = '%Y-%m-%d_%H-%M-%S'  # Cambia esto por el formato deseado

# Leer el archivo CSV
df = pd.read_csv(ruta_archivo)

# Verificar el contenido original
print("Contenido original del archivo:")
print(df.head())

# Convertir la columna 'fecha' a tipo datetime usando el formato especificado
df['fecha'] = pd.to_datetime(df['fecha'], format=formato_original, errors='coerce')

# Verificar si hay fechas que no se pudieron convertir
print(f"Número de fechas no convertidas: {df['fecha'].isna().sum()}")

# Cambiar el formato de la fecha al formato deseado
df['fecha'] = df['fecha'].dt.strftime(formato_deseado)

# Guardar el archivo con el nuevo formato
df.to_csv(ruta_archivo, index=False)

print("Formato de fechas actualizado en el archivo CSV.")
```

```python
#AJUSTAR FORMATO DE FECHAS EN ARCHIVOS CSV DENTRO DE CARPETA


import pandas as pd
import os

# Ruta de la carpeta que contiene los archivos CSV
carpeta_csv = 'C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04/Corriente_Avg'  # Cambia esto por tu ruta

# Especificar el formato original de las fechas y el formato deseado
formato_original = '%Y/%m/%d %H:%M:%S'  # Cambia esto por el formato real
formato_deseado = '%Y-%m-%d_%H-%M-%S'  # Cambia esto por el formato deseado

# Listar todos los archivos CSV en la carpeta
archivos_csv = [f for f in os.listdir(carpeta_csv) if f.endswith('.csv')]

for archivo in archivos_csv:
    ruta_archivo = os.path.join(carpeta_csv, archivo)
    
    # Leer el archivo CSV
    df = pd.read_csv(ruta_archivo)

    # Verificar el contenido original
    print(f"Contenido original del archivo '{archivo}':")
    print(df.head())

    # Convertir la columna 'fecha' a tipo datetime usando el formato especificado
    df['fecha'] = pd.to_datetime(df['fecha'], format=formato_original, errors='coerce')

    # Verificar si hay fechas que no se pudieron convertir
    print(f"Número de fechas no convertidas en '{archivo}': {df['fecha'].isna().sum()}")

    # Cambiar el formato de la fecha al formato deseado
    df['fecha'] = df['fecha'].dt.strftime(formato_deseado)

    # Guardar el archivo con el nuevo formato
    df.to_csv(ruta_archivo, index=False)

    print(f"Formato de fechas actualizado en el archivo '{archivo}'.")

print("Todos los archivos CSV han sido procesados.")
```

```python
#AJUSTA EL FORMATO DE FECHA DE ARCHIVOS CSV QUE ESTAN DENTRO DE SUBCARPETAS DENTRO DE UNA CARPETA PRINCIPAL


import pandas as pd
import os

# Ruta de la carpeta principal que contiene las subcarpetas con archivos CSV
carpeta_principal = 'C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04/'  # Cambia esto por tu ruta

# Especificar el formato original de las fechas y el formato deseado
formato_original = '%Y/%m/%d %H:%M:%S'  # Cambia esto por el formato real
formato_deseado = '%Y-%m-%d_%H-%M-%S'  # Cambia esto por el formato deseado

# Listar todas las subcarpetas en la carpeta principal
subcarpetas = [f.path for f in os.scandir(carpeta_principal) if f.is_dir()]

for subcarpeta in subcarpetas:
    # Listar todos los archivos CSV en la subcarpeta
    archivos_csv = [f for f in os.listdir(subcarpeta) if f.endswith('.csv')]

    for archivo in archivos_csv:
        ruta_archivo = os.path.join(subcarpeta, archivo)

        try:
            # Leer el archivo CSV
            df = pd.read_csv(ruta_archivo)

            # Verificar si la columna 'fecha' existe
            if 'fecha' not in df.columns:
                print(f"El archivo '{archivo}' no contiene la columna 'fecha'.")
                continue

            # Verificar el contenido original
            print(f"Contenido original del archivo '{archivo}' en la subcarpeta '{subcarpeta}':")
            print(df.head())

            # Convertir la columna 'fecha' a tipo datetime usando el formato especificado
            df['fecha'] = pd.to_datetime(df['fecha'], format=formato_original, errors='coerce')

            # Verificar si hay fechas que no se pudieron convertir
            fechas_no_convertidas = df['fecha'].isna().sum()
            if fechas_no_convertidas > 0:
                print(f"Número de fechas no convertidas en '{archivo}': {fechas_no_convertidas}")

            # Cambiar el formato de la fecha al formato deseado
            df['fecha'] = df['fecha'].dt.strftime(formato_deseado)

            # Guardar el archivo con el nuevo formato
            df.to_csv(ruta_archivo, index=False)

            print(f"Formato de fechas actualizado en el archivo '{archivo}'.")
        
        except Exception as e:
            print(f"Error procesando el archivo '{archivo}': {e}")

print("Todos los archivos CSV en las subcarpetas han sido procesados.")
```

CONVERTIDOR DE ARCHIVOS TXT A CSV

```python
#CONVERTIDOR DE ARCHIVO TXT A CSV


import pandas as pd

# Ruta del archivo TXT que deseas convertir
ruta_txt = 'C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04/Corriente_Avg/2024-09-04 00.txt'  # Cambia esto por el nombre real del archivo

# Leer el archivo TXT
with open(ruta_txt, 'r') as f:
    lines = f.readlines()

# Procesar las líneas y crear un DataFrame
data = []
for line in lines:
    # Separar la línea en fecha y valor
    fecha, valor = line.strip().split(',')
    
    # Formatear la fecha eliminando el punto decimal en los segundos
    formatted_fecha = fecha.replace(" ", "_").replace("-", "-")
    parts = formatted_fecha.split("_")
    seconds = parts[-1].split(".")[0]  # Eliminar el decimal
    formatted_fecha = "_".join(parts[:-1] + [seconds])  # Reconstruir la fecha

    data.append([formatted_fecha.strip(), float(valor.strip())])

# Crear un DataFrame
df = pd.DataFrame(data, columns=['fecha', 'valor'])

# Guardar el DataFrame como archivo CSV
ruta_csv = ruta_txt.replace('.txt', '.csv')
df.to_csv(ruta_csv, index=False)

print(f"Archivo '{ruta_txt}' convertido a '{ruta_csv}' con formato de fecha personalizado.")
```

```python
#CONVERTIDOR DE ARCHIVOS TXT A CSV QUE ESTAN EN UNA CARPETA, COMO RESULTADO TE ENTREGA UN CSV CONCATENADO DE LOS TXT QUE ESTAN DENTRO DE LA CARPETA ELEGIDA


import pandas as pd
import os

# Especificar la carpeta que contiene los archivos TXT
nombre_carpeta = "C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04/Corriente_linea1"

# Obtener la ruta completa de la carpeta
carpeta = os.path.join(os.getcwd(), nombre_carpeta)

# Listar todos los archivos TXT en la carpeta especificada
archivos_txt = [f for f in os.listdir(carpeta) if f.endswith('.txt')]
print("Archivos encontrados:", archivos_txt)

# Crear una lista para almacenar DataFrames
dataframes = []

for archivo in archivos_txt:
    # Leer el archivo TXT
    ruta_archivo = os.path.join(carpeta, archivo)
    
    # Leer el archivo y procesar cada línea
    with open(ruta_archivo, 'r') as f:
        lines = f.readlines()
    
    # Procesar las líneas y crear un DataFrame
    data = []
    for line in lines:
        # Separar la línea en fecha y valor
        fecha, valor = line.strip().split(',')
        # Formatear la fecha eliminando el punto decimal en los segundos
        formatted_fecha = fecha.replace(" ", "_").replace("-", "-")
        parts = formatted_fecha.split("_")
        seconds = parts[-1].split(".")[0]  # Eliminar el decimal
        formatted_fecha = "_".join(parts[:-1] + [seconds])  # Reconstruir la fecha
        data.append([formatted_fecha.strip(), float(valor.strip())])
    
    # Crear un DataFrame
    df = pd.DataFrame(data, columns=['fecha', 'valor'])
    
    # Añadir el DataFrame a la lista
    dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_concatenado = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame concatenado como un archivo CSV
ruta_csv_final = os.path.join(carpeta, 'archivo_concatenado.csv')
df_concatenado.to_csv(ruta_csv_final, index=False)

print(f"Todos los archivos TXT se han concatenado y guardado como 'archivo_concatenado.csv' en la carpeta especificada.")
```

```python
#CONVERTIR TODOS LOS ARCHIVOS TXT A CSV QUE ESTAN DENTRO DE SUBCARPETOS EN UNA CARPETA PRINCIPAL, EN UN SOLO ARCHIVO CONCATENADO CORRESPONDIENTE A CADA CARPETA


import pandas as pd
import os

# Especificar la carpeta principal que contiene las subcarpetas con archivos TXT
carpeta_principal = "C:/Users/cesar chavez/Desktop/proyecto/CARPETAS DE DATOS/2024-09-04"  # Cambia esto por la ruta de tu carpeta principal

# Listar todas las subcarpetas en la carpeta principal
subcarpetas = [f.path for f in os.scandir(carpeta_principal) if f.is_dir()]

for subcarpeta in subcarpetas:
    # Inicializar una lista para almacenar los DataFrames de esta subcarpeta
    dataframes = []
    
    # Listar todos los archivos TXT en la subcarpeta
    archivos_txt = [f for f in os.listdir(subcarpeta) if f.endswith('.txt')]
    
    for archivo in archivos_txt:
        # Leer el archivo TXT
        ruta_archivo = os.path.join(subcarpeta, archivo)
        
        # Leer el archivo y procesar cada línea
        with open(ruta_archivo, 'r') as f:
            lines = f.readlines()
        
        # Procesar las líneas y crear un DataFrame
        data = []
        for line in lines:
            # Separar la línea en fecha y valor
            fecha, valor = line.strip().split(',')
            # Formatear la fecha eliminando el punto decimal en los segundos
            formatted_fecha = fecha.replace(" ", "_").replace("-", "-")  # Reemplazar espacio con guion bajo
            parts = formatted_fecha.split("_")
            seconds = parts[-1].split(".")[0]  # Eliminar el decimal
            formatted_fecha = "_".join(parts[:-1] + [seconds])  # Reconstruir la fecha
            data.append([formatted_fecha.strip(), float(valor.strip())])
        
        # Crear un DataFrame y agregarlo a la lista
        df = pd.DataFrame(data, columns=['fecha', 'valor'])
        dataframes.append(df)
        
        print(f"Archivo '{archivo}' en la subcarpeta '{subcarpeta}' procesado.")
    
    # Concatenar todos los DataFrames de la subcarpeta en uno solo
    if dataframes:  # Verificar que hay archivos para concatenar
        df_final = pd.concat(dataframes, ignore_index=True)
        
        # Guardar el DataFrame concatenado como archivo CSV en la subcarpeta
        ruta_csv_final = os.path.join(subcarpeta, 'archivo_concatenado.csv')
        df_final.to_csv(ruta_csv_final, index=False)
        
        print(f"Todos los archivos TXT en la subcarpeta '{subcarpeta}' han sido concatenados y guardados como '{ruta_csv_final}'.")
```
