PRIMERO INICIAREMOS EXPORTANDO LOS DATOS DE DATOS GENERALES EN NUESTRA CARPETA
 
DATOS MANIPULACIÓN

```python
import os
import pandas as pd

name = 'Datos_generales'

# Definir la ruta base hacia 'Datos_generales'
ruta_base = os.path.join('..', 'Datos_generales')  # Carpeta externa

# Crear carpeta para guardar los resultados
output_directory = f"{name}_csv"
os.makedirs(output_directory, exist_ok=True)  # Crear la carpeta si no existe

def concat_txt_to_csv(base_directory, output_directory):
    # Iterar sobre todas las subcarpetas
    for dirpath, dirnames, _ in os.walk(base_directory):
        for dirname in dirnames:
            subdirectory = os.path.join(dirpath, dirname)
            output_csv = f"{dirname}.csv"  # Nombre del archivo CSV de salida basado en la subcarpeta
            
            # Lista para almacenar los datos de los archivos de la subcarpeta
            data_frames = []
            
            # Iterar sobre los archivos en la subcarpeta
            for file in os.listdir(subdirectory):
                if file.endswith(".txt"):
                    file_path = os.path.join(subdirectory, file)
                    print(f"Leyendo archivo: {file_path}")
                    
                    # Leer cada archivo TXT
                    try:
                        with open(file_path, 'r') as f:
                            for line in f:
                                # Dividir la línea por la coma y quitar espacios
                                parts = line.strip().split(',')
                                if len(parts) == 2:  # Asegurarse de que haya exactamente dos partes
                                    # Crear un DataFrame a partir de las partes
                                    df = pd.DataFrame([[parts[0].strip(), float(parts[1].strip())]], columns=['fecha', 'Corriente_linea1'])
                                    data_frames.append(df)
                    except Exception as e:
                        print(f"Error al leer {file_path}: {e}")

            # Concatenar todos los archivos de la subcarpeta en uno
            if data_frames:  # Asegurarse de que hay DataFrames para concatenar
                combined_df = pd.concat(data_frames, ignore_index=True)
                
                # Guardar el DataFrame combinado en un CSV en la carpeta de resultados
                combined_csv_path = os.path.join(output_directory, output_csv)
                combined_df.to_csv(combined_csv_path, index=False)
                print(f"Archivos de la subcarpeta '{dirname}' concatenados y guardados en: {combined_csv_path}")
            else:
                print(f"No se encontraron archivos .txt en la subcarpeta '{dirname}'.")

# Llamar a la función para concatenar los archivos
concat_txt_to_csv(ruta_base, output_directory)
```
