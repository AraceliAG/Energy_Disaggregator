```python
#AGREGAMOS EL DISPOSITIVO EL CUAL SE REQUIERA VISAULIZAR


dispo = 'Datos_generales'
variable = 'Corriente_linea1'




dispositivo = f"Datos_generales_{variable}" #AGREGAMOS EL DISPOSITIVO EL CUAL SE REQUIERA VISAULIZAR

import matplotlib.pyplot as plt
import pandas as pd
import os

# Define la ruta base donde se encuentra la carpeta de datos de entrenamiento
# Aseg√∫rate de usar la ruta absoluta si ejecutas el c√≥digo en un entorno interactivo
ruta_base = '..'



# Carga los datos del archivo CSV correspondiente
archivo_csv = os.path.join(ruta_base, f"{dispositivo}.csv")
datos = pd.read_csv(archivo_csv)

# Convertir la columna 'fecha' a tipo datetime si no lo es
if not isinstance(datos['fecha'], pd.DatetimeIndex):
    datos['fecha'] = pd.to_datetime(datos['fecha'], format='%Y-%m-%d_%H-%M-%S', errors='coerce')

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(datos['fecha'], datos['Corriente_linea1'], marker='o', linestyle='-', color='b')
plt.title(f"Corriente del {dispositivo} a lo Largo del Tiempo")
plt.xlabel('Fecha')
plt.ylabel('Corriente (A)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()

print(f"HORA DE INICIO: {min(datos['fecha'])}")
print(" ")
print(f"HORA DE FINAL: {max(datos['fecha'])}")
```

EN EL BLOQUE ANTERIOR PODEMOS OBSERVAR EL COMPORTAMIENTO GENERAL DE LOS DATOS ANTES DE SEGREGAR,
AHORA BIEN EN EL RANGO DE TIEMPO PODEMOS SELECCIONAR LA SE√ëAL QUE DESEAMOS ESTO A PARTIR DE LAS HORAS QUE NOS MUESTRA
AL FINAL DEL BLOQUE ANTERIOR

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import tkinter as tk
from tkinter import messagebox

# Definir funci√≥n para filtrar los datos y graficar
def filtrar_y_graficar():
    try:
        # Convertir las entradas a datetime aceptando el formato 'YYYY-MM-DD HH:MM:SS'
        hora_inicio = pd.to_datetime(hora_inicio_entry.get(), format='%Y-%m-%d %H:%M:%S')
        hora_final = pd.to_datetime(hora_final_entry.get(), format='%Y-%m-%d %H:%M:%S')

        # Filtrar los datos en base a las fechas proporcionadas
        datos_filtrados = datos[(datos['fecha'] >= hora_inicio) & (datos['fecha'] <= hora_final)]

        if datos_filtrados.empty:
            messagebox.showinfo("Error", "No se encontraron datos en el rango proporcionado.")
        else:
            # Graficar los datos filtrados
            plt.figure(figsize=(12, 6))
            plt.plot(datos_filtrados['fecha'], datos_filtrados['Corriente_linea1'], marker='o', linestyle='-', color='b')
            plt.title(f"Corriente del {dispositivo} a lo Largo del Tiempo (Filtrado)")
            plt.xlabel('Fecha')
            plt.ylabel('Corriente (A)')
            plt.xticks(rotation=45)
            plt.grid()
            plt.tight_layout()
            plt.show()

            # Cerrar la ventana despu√©s de graficar
            ventana.destroy()

    except ValueError as e:
        messagebox.showerror("Error", f"Formato de fecha/hora inv√°lido. Debe ser 'YYYY-MM-DD HH:MM:SS'.\n\n{e}")
        return

# Convertir la columna 'fecha' a tipo datetime usando el formato adecuado del archivo CSV
datos['fecha'] = pd.to_datetime(datos['fecha'], format='%Y-%m-%d_%H-%M-%S', errors='coerce')

# Crear ventana con tkinter
ventana = tk.Tk()
ventana.title("Seleccionar Hora de Inicio y Final")
ventana.geometry("400x200")

# Etiquetas y campos para ingresar hora de inicio y final
tk.Label(ventana, text="Hora de Inicio (YYYY-MM-DD HH:MM:SS)").pack(pady=5)
hora_inicio_entry = tk.Entry(ventana, width=30)
hora_inicio_entry.pack(pady=5)

tk.Label(ventana, text="Hora de Final (YYYY-MM-DD HH:MM:SS)").pack(pady=5)
hora_final_entry = tk.Entry(ventana, width=30)
hora_final_entry.pack(pady=5)

# Bot√≥n para filtrar y graficar
filtrar_btn = tk.Button(ventana, text="Filtrar y Graficar", command=filtrar_y_graficar)
filtrar_btn.pack(pady=20)

# Iniciar el loop de la ventana
ventana.mainloop()
```
